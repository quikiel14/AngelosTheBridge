## Ejercicio 1
Somos una empresa que tiene varias tiendas de electrodomésticos. Necesitamos un programa para manejar las tiendas, ver las ventas que han tenido, cuántos empleados hay, etc... Para ello vamos a modelizar la tienda en una clase, que tendrá las siguientes características:
* Nombre clase: "Tienda"
* Atributos comunes:
    * Tipo: "Electrodomésticos"
    * Abierta: True
* Atributos propios de cada tienda:
    * Nombre - String
    * Dirección - String
    * Número de empleados - int
    * Ventas últimos 3 meses - Lista de 3 numeros
* Método para calcular las ventas de todos los meses, que devuelva un número con todas las ventas.
* Método que calcula la media de ventas de los últimos meses, por empleado
* Método que devuelve en un string el nombre de la tienda, junto con su dirección.
* Método que obtiene las ventas del último mes.
* Método para dar una proyección de las ventas en caso de haber invertido X dinero en marketing en esa tienda. Siendo X el parámetro de entrada. Si X es menor de 1000, las ventas de los ultimos 3 meses hubiesen sido de (1.2 x ventas), si es mayor o igual de 1000, las ventas hubiesen sido de (1.5 x venta). El método reescribe el atributo de ventas en los últimos 3 meses, y además devuelve las ventas con el aumento.

Se pide:
1. Implementar la clase "Tienda"

4. Imprime por pantalla los nombres de las tiendas cuya dirección lleve el string "Avenida"
5. Seguro que ya has hecho este apartado... Documenta la clase :)

```python
class Tienda:
    """
    Clase que representa una tienda de electrodomésticos.
    
    Atributos comunes:
            * tienda_tipo (str): Tipo de negocio, siempre "Electrodomésticos".
            * tienda_abierta (bool): Indica si la tienda está abierta.
    
    Atributos propios de cada tienda:
            * Nombre (str): Nombre de la tienda.
            * Direccion (str): Dirección de la tienda.
            * Num_empl (int): Número de empleados en la tienda.
            * ventas_3_meses (list): Lista con las ventas de los últimos 3 meses.
    """
    tienda_tipo = "Electrodomésticos"
    tienda_abierta = True
    ventas_3_meses = [None,None,None]
    Direccion = None
    Nombre = None
    Num_empl = None
    inversion = None


    def __init__(self,Nombre:str, Direccion:str, Num_empl:int, ventas_3_meses:list):
        self.Nombre = Nombre
        self.Direccion = Direccion
        self.Num_empl = Num_empl
        self.ventas_3_meses = ventas_3_meses
        


    def VentasTodos (self,):
        """Calcula el total de ventas de los últimos 3 meses."""
        cuenta_precios = 0
        for i in self.ventas_3_meses:
            cuenta_precios = cuenta_precios + i
        return cuenta_precios
    
    def MediaVentas(self,):
        """Calcula la media de ventas por empleado."""
        return self.VentasTodos() / self.Num_empl
    
    def StringNombreDirr(self,) -> str:
        """Devuelve un string con el nombre y dirección de la tienda."""
        return self.Nombre +" "+ self.Direccion
    
    def Ultimo_Mes (self,):
        """Obtiene las ventas del último mes."""
        return self.ventas_3_meses[-1]
    
    def Marketing (self, inversion:int,):
        """Calcula la proyección de ventas si se invierte en marketing."""
        lista_pro = []
        self.inversion = inversion
        print(f"La lista original :",self.ventas_3_meses)
        if self.inversion < 1000 :
            for i in self.ventas_3_meses:
                lista_pro.append(i* 1.2)
                

        else:
            for i in self.ventas_3_meses:
               lista_pro.append(i * 1.5)
        print(f"La lista de proyección :",lista_pro)
        

# Crear tres tiendas
Lidl = Tienda("Lild","Masustegui 8 ",300,[23,34,54],)
MediaMarkt = Tienda("MediaMarkt","Avenida 32 ",32,[2,4,5])
IKEA_e = Tienda("IKEA_e","Ronda 2 ",40,[3,66,15])

# Comprobar funcionalidades en una tienda
print(Lidl.__dict__)
print(Lidl.VentasTodos())
print(Lidl.MediaVentas())
print(Lidl.StringNombreDirr())
print(Lidl.Ultimo_Mes())
Lidl.Marketing(1500)

# Calcular ventas del último mes de todas las tiendas
lista = []
for i in Lidl.ventas_3_meses,MediaMarkt.ventas_3_meses,IKEA_e.ventas_3_meses:
    lista.append(i[-1])

print(lista)

# Imprimir nombres de tiendas cuya dirección contiene "Avenida"
print("Tiendas ubicadas en una Avenida:")
for tienda in [Lidl, MediaMarkt, IKEA_e]:
    if "Avenida" in tienda.Direccion:
        print(tienda.Nombre)
    





    


    


```

## Ejercicio 2
En este ejercicio vamos a implementar una clase *Perro* en Python. La clase tiene las siguientes características:
* Cosas que sabemos seguro que tiene un perro
    * Tiene 4 patas
    * 2 orejas
    * 2 ojos
    * Una velocidad de 0. Por defecto, el perro está parado
* Cuando se inicialice:
    * El perro será de una determinada raza
    * Por defecto tendrá pelo "Marrón", a no ser que se diga lo contrario.
    * Por defecto no tendrá dueño, a no ser que se diga lo contrario.
    
* Dispondrá también de un método llamado andar, que tiene un argumento de entrada (aumento_velocidad). Este valor se le sumará a la velocidad que ya llevaba el perro.
* Necesita otro método (parar), donde pondremos la velocidad a 0.
* Otro método llamado "ladrar", que tendrá un argumento de entrada, y la salida será el siguiente string: "GUAU!" + el argumento de entrada.


Se pide:
* Implementa la clase *Perro*
* Crea un objeto de tipo *Perro*, sin dueño
* Comprueba que están bien todos sus atributos
* Prueba que ande, y comprueba su velocidad
* Páralo
* Documenta la clase *Perro*

```python
class Perro:
    """
    Clase que representa a un perro.

    Atributos comunes:
            * patas (int): Número de patas, siempre 4.
            * orejas (int): Número de orejas, siempre 2.
            * ojos (int): Número de ojos, siempre 2.
            * velocidad (int): Velocidad actual del perro (inicialmente 0).

    Atributos específicos de cada perro:
            * raza (str): Raza del perro.
            * pelo (str): Color del pelo, por defecto "Marrón".
            * duenio (str | None): Nombre del dueño, por defecto None.
    """

    patas = 4
    orejas = 2
    ojos = 2

    def __init__(self, raza: str, pelo: str = "Marrón", duenio: str = None):
        self.raza = raza
        self.pelo = pelo
        self.duenio = duenio
        self.velocidad = 0  # El perro empieza parado

    def andar(self, aumento_velocidad: int):
        """Aumenta la velocidad del perro en una cantidad específica."""
        self.velocidad += aumento_velocidad
        return self.velocidad

    def parar(self):
        """Detiene al perro poniendo su velocidad a 0."""
        self.velocidad = 0
        return self.velocidad

    def ladrar(self, sonido: str):
        """Hace que el perro ladre, agregando un sonido personalizado."""
        return f"GUAU! {sonido}"


# Crear un perro sin dueño
mi_perro = Perro(raza="Chow Chow")

# Comprobar sus atributos
print(mi_perro.__dict__)

# Hacer que ande y comprobar su velocidad
print("Velocidad después de andar:", mi_perro.andar(10))

# Pararlo
print("Velocidad después de parar:", mi_perro.parar())

# Hacerlo ladrar
print(mi_perro.ladrar("Alguien en la puerta!"))
```
