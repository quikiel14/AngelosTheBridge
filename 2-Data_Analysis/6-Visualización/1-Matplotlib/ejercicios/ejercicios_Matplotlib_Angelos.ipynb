2. Sabemos que podemos pintar gráficas de dos formas: la figura incluye los ejes o teniendo figura y ejes por separado.

Usando solo una figura, usa numpy para los valores del eje X entre 0 y 5. 

Pinta dos gráficas en dos cajas distintas, a la izquierda una recta con pendiente positiva de 3 que pase por (0,0) y a la derecha una recta con pendiente negativa de 3 que pase por (0,-5).

```python
y1 = 3*x 
y2 = -3*x -5 



fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))

# Gráfica izquierda
ax1.plot(x, y1, color='blue')
ax1.set_title("Pendiente +3, pasa por (0, 0)")
ax1.grid(True)

ax2.plot(x, y2, color='red')
ax2.set_title("Pendiente -3, pasa por (0, -5)")
ax2.grid(True)


plt.tight_layout()
plt.show()


```

3. Fija el eje X entre 0 y 5 y el eje Y entre -15 y 15

```python
y1 = 3*x 
y2 = -3*x -5 

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))


ax1.plot(x, y1, color='blue')
ax1.set_title("Pendiente +3, pasa por (0, 0)")
ax1.set(xlim=(0, 5), ylim=(-15, 15))
ax1.grid(True)


ax2.plot(x, y2, color='red')
ax2.set_title("Pendiente -3, pasa por (0, -5)")
ax1.set(xlim=(0, 5), ylim=(-15, 15))
ax2.grid(True)

plt.tight_layout()
plt.show()
```

4. Llama al eje X "eje X", al eje Y "eje Y" y pon de títulos "recta sube" y "recta baja". Muestra dos etiquetas de ejeX pero solo una de eje Y.

```python
y1 = 3*x 
y2 = -3*x -5 

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))


ax1.plot(x, y1, color='blue')
ax1.set_title("recta sube")
ax1.set(xlim=(0, 5), ylim=(-15, 15),
        xlabel = "eje X",
        ylabel = "eje Y")
ax1.grid(True)


ax2.plot(x, y2, color='red')
ax2.set_title("recta baja")
ax1.set(xlim=(0, 5), ylim=(-15, 15),
        xlabel = "eje X",
        ylabel = "eje Y")
ax2.grid(True)

plt.tight_layout()
plt.show()
```

15. Otro departamento B ha vendido en esos años 2000, 3100, 5000, 4000, 6000 unidades. Incluye esa columna en el DataFrame y pinta en la misma gráfica las dos líneas. B es una línea punteada y roja. Muestra una leyenda abajo a la derecha.

```python
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use('seaborn-v0_8-white')

data = {
    'Year': [2015, 2016, 2017, 2018, 2019],
    'Sold_items_A': [1000, 3500, 4000, 5500, 7000],
    'Sold_items_B': [2000, 3100, 5000, 4000, 6000]
}


df = pd.DataFrame(data)


plt.figure(figsize=(8, 5))

plt.plot(df['Year'], df['Sold_items_A'], linestyle='--', color='green', marker='o', label='Departamento A')


plt.plot(df['Year'], df['Sold_items_B'], linestyle=':', color='red', marker='s', label='Departamento B')


plt.title("Comparativa de ventas anuales")
plt.xlabel("Año")
plt.ylabel("Unidades vendidas")


plt.legend(loc='lower right')

plt.grid(True)

plt.show()
```

17. Prueba a cambiar kind a 'pie'

```python
import pandas as pd
import matplotlib.pyplot as plt


plt.style.use('seaborn-v0_8-white')

data = {
    'Year': [2015, 2016, 2017, 2018, 2019],
    'Sold_items_A': [1000, 3500, 4000, 5500, 7000],
    'Sold_items_B': [2000, 3100, 5000, 4000, 6000]
}

df = pd.DataFrame(data)
df.set_index('Year', inplace=True)
df['Sold_items_A'].plot(kind='pie', autopct='%1.1f%%', figsize=(6, 6), title='Distribución de ventas - Dpto. A')

# plt.xlabel("Año")
# plt.ylabel("Unidades vendidas")
plt.ylabel("")

plt.grid(True)
plt.show()
```

18. Prueba a quitar la leyenda incluyendo legend igual a False, añade labels y quita la etiqueta en Y

```python
import pandas as pd
import matplotlib.pyplot as plt


plt.style.use('seaborn-v0_8-white')

data = {
    'Year': [2015, 2016, 2017, 2018, 2019],
    'Sold_items_A': [1000, 3500, 4000, 5500, 7000],
    'Sold_items_B': [2000, 3100, 5000, 4000, 6000]
}

df = pd.DataFrame(data)

labels = df['Year']

df['Sold_items_A'].plot(
    kind='pie',
    labels=labels,
    autopct='%1.1f%%',
    figsize=(6, 6),
    title='Distribución de ventas - Departamento A',
    legend=False
)

plt.ylabel("")

plt.show()
```

19. Vamos a pintar un histograma de unas alturas generadas aleatoriamente. Crea 300 alturas entre 150 y 190 cm.

Pista: https://pandas.pydata.org/pandas-docs/version/0.23.4/generated/pandas.DataFrame.plot.html

```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn-v0_8-white')

alturas = np.random.uniform(150, 190, 300)

plt.figure(figsize=(8, 5))
plt.hist(alturas, bins=20, color='skyblue', edgecolor='black')

plt.title("Distribución de alturas (150 cm - 190 cm)")
plt.xlabel("Altura (cm)")
plt.ylabel("Frecuencia")

plt.grid(True)
plt.show()
```
